```python
ipl_df2 = ipl_df.rename(columns = {'Player':'player',
                               'Base Price':'base_price',
                               'TYPE':'type', 
                               'COST IN ₹ (CR.)':'cost_inr', 
                               'Cost IN $ (000)':'cost_usd',
                               '	2021 Squad	':'2021_team',
                               'Team':'2022_team'})
ipl_df2.head()
```

```python
ipl_df3 = ipl_df2.drop(['cost_usd'],axis = 1)
ipl_df3.head()
```

```python
mysql = lambda q:sqldf(q,globals())
```

```python
mysql(""" select * from ipl_df3""")
```

## Name top 3 batsman who got paid the most

```python
mysql(""" select player,cost_inr from ipl_df3 where type = 'BATTER' order by cost_inr desc limit 3""")
```

## Name top 5 bowlers who get paid the most

```python
mysql(""" select player, cost_inr from ipl_df3 
where type = 'BOWLER'
order by cost_inr desc
limit 5""")
```

## Name top 5 All-rounders who get paid the most

```python
mysql(""" select player, cost_inr from ipl_df3
where type = 'ALL-ROUNDER'
order by cost_inr desc
limit 5""")
```

## Name 5 lowest paid wicket-keeper

```python
mysql("""select player, cost_inr from ipl_df3
where type = 'WICKETKEEPER' and cost_inr is not null
order by cost_inr asc
limit 5""")
```

## What is the Average pay for Batsman, Bowler, All-rounder, Wicket-keeper

```python
mysql("""select type, avg(cost_inr) as average_price from ipl_df3
group by type
order by avg(cost_inr) desc""")
```

## List of retained players with team name and salary

```python
mysql("""select player,cost_inr from ipl_df3 
where base_price = 'Retained'
order by cost_inr desc
limit 5""")
```
